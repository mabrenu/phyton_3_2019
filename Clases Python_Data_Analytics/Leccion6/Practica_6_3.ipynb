```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
y=(np.arange(0,1000,10))
```

```python
x= y**2
```

```python
plt.plot(x, 'k--', label='Default')
plt.plot(y, 'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
y2= -x**2 + 500 *x
```

```python
plt.plot(x, 'k--', label='Default')
plt.plot(y, 'k-', drawstyle='steps-post', label='steps-post')
plt.plot(y2,'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```

```python
plt.plot(x, 'k--', label='Default')
plt.plot(y, 'k-', drawstyle='steps-post',color='b', label='steps-post')
plt.plot(y2,'k-', drawstyle='steps-post', label='steps-post')
plt.legend(loc='best')
```
